```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data
df = pd.read_csv('/content/gender_classification_dataset.csv')

# Take the top 10 rows
data = df.head(10)

# Extract columns for Name & Height(cm)
x = data['Name']
y = data['Height (cm)']



# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(x, y, color='pink', edgecolor='black')

# Rotate x-axis ticks for clarity
plt.xticks(rotation=45)

# Add titles and labels
plt.title('Height Chart', fontsize=14)
plt.xlabel('Name', fontsize=12)
plt.ylabel('Height(cm)', fontsize=12)

# Improve layout
plt.tight_layout()

# Show the chart
plt.show()
```
